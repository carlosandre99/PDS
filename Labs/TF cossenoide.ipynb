```python
def plota(x):
    X=fft(x)/len(x)
    figure(figsize=(15,4))
    subplot(1,2,1)
    plot(x)
    xlabel('n')
    subplot(1,2,2)
    stem(abs(X),linefmt='C0-',markerfmt='C0o',basefmt='C0-')
    xlabel('k')
    show()
```

Considere um sinal $x[n]$, uma cossenoide de frequencia $w_x$, com 40 valores de n=0 a 39; <br>$x[n]=cos(w_x n)$;

Deseja-se visualizar o espectro de x para valores de $w$ entre 0 e $2\pi$, com 40 valores, em função de cada compartimento (*bin*). Plotar o sinal $x[n]$ e a magnitude da transformada de fourier $X(e^{jw})$ para os seguintes casos, quando $x[n]$ tem:

1. exatamente 1 ciclo
2. exatamente 1,5 ciclo
3. entre 1,5 e 2 ciclos
4. exatamente 10 ciclos
5. entre 10 e 11 ciclos

8. exatamente 21 ciclos

Responder para cada uns dos itens (1 a 8) em qual *bin* (compartimento, valor) dos primeiros 20 valores de frequência ficou concentrada a maior parte da energia do sinal $x[n]$, ou seja, qual valor de $k$ apresentou maior amplitude ?


Implemente função que gera o sinal x de 40 amostras com um número C de ciclos de cossenoide e chama a função plota(x) para plotá-lo.

```python
def generate_cosine_signal(C):
    N = 40
    n = np.arange(N)
    w = 2 * np.pi * C / N  
    x = np.cos(w * n)
    return x

def plotaciclos(C):
    x = generate_cosine_signal(C)
    plota(x)
```
