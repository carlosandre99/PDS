1) Considere a sequência teórica: $x[n]=1$, para $n=0,1,2,3$, e $x[n]=0$, para todos os outros valores de $n$ ($-\infty ... +\infty)$.

2) Calcule a transformada de fourier da sequencia $x[n]$, $n=0$ a 199, obtendo 200 valores na faixa de frequência entre 0 e $2\pi$.

```python
x = [1, 1, 1, 1] + [0] * 196
X = fft(x)
freqs = arange(0, 2*pi, 2*pi/200)
plot(freqs, abs(X))
xlabel('Frequência (rad/s)')
ylabel('Magnitude')
show()
```

3) Calcule a DFT desta sequencia, para n=0 a 3.

```python
dft_3 = fft(x[:4])
```

4) Calcule a DFT desta sequencia, para n=0 a 7.

```python
dft_7 = fft(x[:8])
```

5) Calcule a DFT desta sequencia, para n=0 a 15.

```python
dft_15 = fft(x[:16])
```

6) Plote no mesmo eixo (abscissa de zero a 2$\pi$), todas as 4 transformadas. usando os seguintes simbolos: 2) linha continua, 3) 'o', 4) '+', 5) 'x' .

```python
freqs = arange(0, 2*pi, 2*pi/200)
freqs_3 = arange(0, 2*pi, 2*pi/4)
freqs_7 = arange(0, 2*pi, 2*pi/8)
freqs_15 = arange(0, 2*pi, 2*pi/16)

figure()
plot(freqs, abs(X), label='200', linestyle='-')
plot(freqs_3, abs(dft_3), 'o', label='4')
plot(freqs_7, abs(dft_7), '+', label='8')
plot(freqs_15, abs(dft_15), 'x', label='16')
xlabel('Frequência (rad/s)')
ylabel('Magnitude')
legend()
show()
```

7) Comente o resultado.

É possivel observar a DFT para diferentes tamanhos de janela: 4, 8 e 16, além do tamanho total do sinal (200). Isso é feito para comparar como a resolução da frequência muda com o tamanho da janela.Em geral, um tamanho de janela maior resultará em uma maior resolução de frequência, pois mais pontos de dados permitem uma estimativa mais precisa das frequências presentes no sinal. Por outro lado, um tamanho de janela menor resultará em uma maior resolução temporal (podemos detectar mudanças no sinal mais rapidamente), mas a custo de uma menor resolução de frequência.

8) Gere um vetor x1, uma senoide de 32 pontos, com 2 ciclos completos.

```python
x1 = sin(linspace(0, 4*pi, 32))
stem(x1, linefmt='bo-', markerfmt='bo', basefmt='b-')
```

9) Gere vetor x2 acrescentando zeros ao final do vetor x de forma a duplicar seu tamanho.


```python
x2 = concatenate((x1, zeros(32)))
stem(x2, linefmt='b', markerfmt='b', basefmt='b')

```

10) Gere um vetor x3 com tamanho 128 acrescentando mais zeros


```python
x3 = concatenate((x1, zeros(96)))
stem(x3, linefmt='b', markerfmt='b', basefmt='b')

```

11) Plote no mesmo eixo de frequencia (de 0 a 2pi) o valor absoluto das DFTs X1, X2 e X3 e comente o resultado.
(simbolos X1: 'o', X2: '+', X3: '-'

```python
X1 = fft(x1)
X2 = fft(x2)
X3 = fft(x3)

f1 = arange(0, 2*pi, 2*pi/32)
f2 = arange(0, 2*pi, 2*pi/64)
f3 = arange(0, 2*pi, 2*pi/128)

figure()
plot(f1, abs(X1), 'o', label='32')
plot(f2, abs(X2), '+', label='64')
plot(f3, abs(X3), '-', label='128')
xlabel('Frequência (rad/s)')
ylabel('Magnitude')
legend()
show()
```

12) Comente o resultado.

No gráfico, deve-se observar que todos os três sinais têm picos nas mesmas frequências, pois todos são baseados no mesmo sinal original. No entanto, os sinais com zero padding (x2 e x3) terão uma maior precisão na resolução de frequência devido aos pontos adicionais na FFT. A adição de zeros não muda as frequências presentes no sinal, mas faz com que a FFT tenha mais pontos, o que pode tornar o espectro de frequência mais definido.
