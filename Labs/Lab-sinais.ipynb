## 1) Gerar os seguintes sinais e visualiza-los (comandos plot e stem). 

```python
from pylab import *
```

### a) impulso unitário, para valores de n (abscissa) entre -10 e 10 (stem), inclusive.

```python
n = arange(-10, 11)
impulse = where(n == 0, 1, 0)
subplot(2, 1, 2)
stem(n, impulse, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### b) degrau unitário, para n entre -10 e 10 (stem), inclusive.


```python
n = arange(-10, 11)
unit_step = where(n >= 0, 1, 0)
subplot(2, 1, 2)
stem(n, unit_step, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### c) ruido branco gaussiano (média zero, variância unitária) com 40 valores, para n=0 a 39 (stem).

```python
n = arange(0, 40)
mean = 0
std = 1
white_noise = np.random.normal(mean, std, size=len(n))
stem(n, white_noise, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### d) senoide com 4 ciclos, com 90 valores, para n=0 a 89 (stem).

```python
n = arange(0, 90)
frequency = 4 / 90
sine_wave = sin(2 * pi * frequency * n)
stem(n, sine_wave, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### e) cosenoide com 5 ciclos, com 110 valores, para n=0 a 109 (stem). 

```python
n = arange(0, 110)
frequency = 5 / 110
cosine_wave = cos(2 * pi * frequency * n)
stem(n, cosine_wave, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### f) exponencial complexa com 5 ciclos, para n=0 a 69. Parte real (stem) e parte imaginária (plot). Usar subplot para plotar em dois graficos separados: subplot(2,1,1) para o primeiro e subplot(2,1,2) para do segundo.

```python
n = arange(0, 70)
frequency = 5 / 70
complex_exponential = exp(1j * 2 * pi * frequency * n)
real_part = real(complex_exponential)
imaginary_part = imag(complex_exponential)

# Subplot para a parte real 
subplot(2, 1, 1)
stem(n, real_part, basefmt='r', markerfmt='bo', linefmt='b')

# Subplot para a parte imaginária 
subplot(2, 1, 2)
plot(n, imaginary_part, 'g')

tight_layout()  # Para melhor espaçamento entre os subplots
show()
```

### g) senoide com frequência w=0,2 rd ( sin(w*n) ) , para n de 0 a 99 (plot).

```python
n = arange(0, 100)
angular_frequency = 0.2
sine_wave = sin(angular_frequency * n)
plot(n, sine_wave, 'b')
show()
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), com 70 valores, para n de 0 a 69 (plot).

```python
frequency = 60  
sampling_rate = 800  
num_samples = 70  
period = 1 / frequency
n = arange(0, num_samples)
sine_wave = sin(2 * pi * frequency * n / sampling_rate)
plot(n, sine_wave, 'b')
show()
```

### l) exponencial decrescente para n = 0 a 60 (stem).

```python
n = arange(0, 61)
decay_rate = 0.9
exponential_decay = decay_rate ** n
stem(n, exponential_decay, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### m) exponencial oscilante decrescente $x[n]=z^n$, para n=0 a 90, parte real (stem).

```python
n = arange(0, 91)
z_magnitude = 0.9
z_angle = pi / 4  
exponential_oscillating_decay = z_magnitude ** n * cos(z_angle * n)
stem(n, exponential_oscillating_decay, basefmt='r', markerfmt='bo', linefmt='b')
show()
```

### n) exponencial oscilante crescente $x[n]=z^n$, para n=0 a 80, parte imaginaria (stem).

```python
n = arange(0, 81)
z_magnitude = 1.1
z_angle = pi / 3  
exponential_oscillating_rise = z_magnitude ** n * sin(z_angle * n)
stem(n, exponential_oscillating_rise, basefmt='r', markerfmt='bo', linefmt='b')
show()
```
